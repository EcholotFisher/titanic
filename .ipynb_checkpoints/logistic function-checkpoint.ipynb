```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
plt.style.use('seaborn')
```

```python
def logistic(t):
    return 1/(1+np.exp(-t))
```

```python
np.arange(-10,10,0.1)
```

```python
z = np.arange(-5,5,0.1)
sigma_t = [a for a in map(logistic,z)]



fig, ax = plt.subplots()
plt.plot(z,sigma_t)
plt.plot(z,np.repeat(0.5,len(z)),linestyle=':',color='black')
plt.plot(z,np.repeat(1,len(z)),linestyle=':',color='black')
plt.axvline(0,linestyle='-',color='black',ylabel='test')
#plt.ylabel('$\sigma(t)$')
plt.xlabel('t')
```

```python
fig, (ax1,ax2) = plt.subplots(2,1)
ax1.plot(sigma_t,-np.log(sigma_t))
ax1.set_ylabel('$-log(\hat{p})$')
ax1.set_title('y=1')

ax2.plot(sigma_t,-np.log([1-a for a in sigma_t]))
ax2.set_ylabel('$-log((1-\hat{p}))$')
ax2.set_title('y=0')
ax2.set_xlabel('$\hat{p}$')
fig.suptitle('Loss as a function of estimated probability')
#ax1.xlabel('$\hat{p}$')
```
